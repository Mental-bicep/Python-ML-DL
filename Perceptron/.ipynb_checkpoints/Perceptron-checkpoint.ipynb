```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y=make_blobs(n_samples=500,n_features=2,centers=2,random_state=4)
```

```python
print(x.shape)
print(y.shape)
```

```python
plt.style.use("seaborn")
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
z=np.array([1,2,3,4,5,6])
print(sigmoid(z))
```

```python
def predict(x,weights):
    z=np.dot(x,weights)
    pred=sigmoid(z)
    return pred
```

```python
def loss(x,y,weights):
    y_=predict(x,weights)
    cost=np.mean(-y*np.log(y_)-(1-y)*np.log(1-y_))
    return cost
```

```python
def update(x,y,weights,l_rate):
    y_=predict(x,weights)
    dw=np.dot(x.T,y_-y)
    m=x.shape[0]
    weights=weights-l_rate*dw/float(m)
    return weights
```

```python
def train(x,y,l_rate=0.5,max_epochs=100):
    ones=np.ones((x.shape[0],1))
    x=np.hstack((ones,x))
    weights=np.zeros(x.shape[1])
    
    for epoch in range(max_epochs):
        weight=update(x,y,weights,l_rate)
        if(epoch%10==0):
            l=loss(x,y,weights)
#             print("epoch %d loss %0.4f"%(epoch,l))
    return weights
```

```python
print(train(x,y))
```

```python
weights=train(x,y,l_rate=0.8,max_epochs=1000)
print(weights)
```
