```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_=cv2.imread("download (2).jpg")
```

```python
plt.imshow(img_)
```

```python
img_=cv2.cvtColor(img_,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_)
```

```python
img=cv2.resize(img_,(100,100))
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
def drawing(img,title="image"):
    plt.imshow(img,cmap="gray")
    plt.axis("off")
    plt.title(title+str(img.shape))
    plt.show()
drawing(img)
```

```python
drawing(img_gray)
```

```python
def convolution(img,img_filter):
    w=img.shape[0]
    h=img.shape[1]
    f=img_filter.shape[0]
    new_img=np.zeros((w-f+1,h-f+1))
    for row in range(w-f+1):
        for col in range(h-f+1):
            for i in range(f):
                for j in range(f):
                    new_img[row][col]+=img[row+i][col+j]*img_filter[i][j]
                
                if new_img[row][col]>255:new_img[row][col]=255
                if new_img[row][col]<0:new_img[row][col]=0
    return new_img
```

```python
blur_filter=np.ones((3,3))/9.0
```

```python
output=convolution(img_gray,blur_filter)
```

```python
drawing(output)
drawing(img_gray)
```
